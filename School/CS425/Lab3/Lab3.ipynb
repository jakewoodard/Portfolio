```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
from sklearn.cluster import KMeans
from math import floor
```

```python
# Question 1

def func(file_name, k):

    # Read in the image
    image = cv2.imread(file_name)
    # Change color to RGB (from BGR)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    x, y, z = image.shape
    image2 = image.reshape(x*y, z)
    image2.shape
    
    kmeans_cluster = KMeans(n_clusters=k, random_state=42)
    kmeans_cluster.fit(image2)
    cluster_centers = kmeans_cluster.cluster_centers_
    for i in range(0, len(cluster_centers)):
        for j in range(0, len(cluster_centers[i])):
            cluster_centers[i][j] = int(floor(cluster_centers[i][j]))

    cluster_labels = kmeans_cluster.labels_
    return cluster_centers,cluster_labels,x,y,z
```

```python
# Question 2

# Checker Neyland
print("Neyland Stadium\n")
plt.figure(figsize=(10,10))

# ORIGINAL
image = cv2.imread('checkerneyland.jpeg')
plt.subplot(2,2,1),plt.imshow(image)
plt.title('Original'), plt.xticks([]), plt.yticks([])
# K = 4
clusters, labels, X, Y, Z = func('checkerneyland.jpeg', 4)
plt.subplot(2,2,2),plt.imshow(((clusters[labels].reshape(X, Y, Z)) * 255).astype(np.uint8))
plt.title('K = 4'), plt.xticks([]), plt.yticks([])
# K = 8
clusters, labels, X, Y, Z = func('checkerneyland.jpeg', 8)
plt.subplot(2,2,3),plt.imshow(((clusters[labels].reshape(X, Y, Z)) * 255).astype(np.uint8))
plt.title('K = 8'), plt.xticks([]), plt.yticks([])
# K = 16
clusters, labels, X, Y, Z = func('checkerneyland.jpeg', 16)
plt.subplot(2,2,4),plt.imshow(((clusters[labels].reshape(X, Y, Z)) * 255).astype(np.uint8))
plt.title('K = 16'), plt.xticks([]), plt.yticks([])
plt.show()

# Min Kao
print("Min Kao Building\n")
plt.figure(figsize=(10,10))

# ORIGINAL
image = cv2.imread('minkao.jpeg')
plt.subplot(2,2,1),plt.imshow(image)
plt.title('Original'), plt.xticks([]), plt.yticks([])
# K = 4
clusters, labels, X, Y, Z = func('minkao.jpeg', 4)
plt.subplot(2,2,2),plt.imshow(((clusters[labels].reshape(X, Y, Z)) * 255).astype(np.uint8))
plt.title('K = 4'), plt.xticks([]), plt.yticks([])
# K = 8
clusters, labels, X, Y, Z = func('minkao.jpeg', 8)
plt.subplot(2,2,3),plt.imshow(((clusters[labels].reshape(X, Y, Z)) * 255).astype(np.uint8))
plt.title('K = 8'), plt.xticks([]), plt.yticks([])
# K = 16
clusters, labels, X, Y, Z = func('minkao.jpeg', 16)
plt.subplot(2,2,4),plt.imshow(((clusters[labels].reshape(X, Y, Z)) * 255).astype(np.uint8))
plt.title('K = 16'), plt.xticks([]), plt.yticks([])
plt.show()

# Smokey
print("Smokey\n")
plt.figure(figsize=(10,10))

# ORIGINAL
image = cv2.imread('smokey.jpeg')
plt.subplot(2,2,1),plt.imshow(image)
plt.title('Original'), plt.xticks([]), plt.yticks([])
# K = 4
clusters, labels, X, Y, Z = func('smokey.jpeg', 4)
plt.subplot(2,2,2),plt.imshow(((clusters[labels].reshape(X, Y, Z)) * 255).astype(np.uint8))
plt.title('K = 4'), plt.xticks([]), plt.yticks([])
# K = 8
clusters, labels, X, Y, Z = func('smokey.jpeg', 8)
plt.subplot(2,2,3),plt.imshow(((clusters[labels].reshape(X, Y, Z)) * 255).astype(np.uint8))
plt.title('K = 8'), plt.xticks([]), plt.yticks([])
# K = 16
clusters, labels, X, Y, Z = func('smokey.jpeg', 16)
plt.subplot(2,2,4),plt.imshow(((clusters[labels].reshape(X, Y, Z)) * 255).astype(np.uint8))
plt.title('K = 16'), plt.xticks([]), plt.yticks([])
plt.show()

# Ayres Hall
print("Ayres Hall\n")
plt.figure(figsize=(10,10))

# ORIGINAL
image = cv2.imread('ayreshall.jpeg')
plt.subplot(2,2,1),plt.imshow(image)
plt.title('Original'), plt.xticks([]), plt.yticks([])
# K = 4
clusters, labels, X, Y, Z = func('ayreshall.jpeg', 4)
plt.subplot(2,2,2),plt.imshow(((clusters[labels].reshape(X, Y, Z)) * 255).astype(np.uint8))
plt.title('K = 4'), plt.xticks([]), plt.yticks([])
# K = 8
clusters, labels, X, Y, Z = func('ayreshall.jpeg', 8)
plt.subplot(2,2,3),plt.imshow(((clusters[labels].reshape(X, Y, Z)) * 255).astype(np.uint8))
plt.title('K = 8'), plt.xticks([]), plt.yticks([])
# K = 16
clusters, labels, X, Y, Z = func('ayreshall.jpeg', 16)
plt.subplot(2,2,4),plt.imshow(((clusters[labels].reshape(X, Y, Z)) * 255).astype(np.uint8))
plt.title('K = 16'), plt.xticks([]), plt.yticks([])
plt.show()
```
