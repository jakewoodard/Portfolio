```python
import numpy as np
import matplotlib.pyplot as plt
```

### Experiment 1

```python
# generate 50 bipolar vectors of length 100
vector_list = []
for i in range(0,50):
    vector_list.append(list(np.random.choice([-1,1], 100)))
    
# sign func
def sgn(h):
    if h < 0: return -1
    return 1
```

```python
# Create Hopfield network
W = np.zeros((100, 100))
stable_list = [0]*50
prob_list = [0]*50
unprob_list = [0]*50

# imprint weights and test patterns
for p in range(50):
    network_state = vector_list[p].copy()
    for i in range(100):
        h_i = 0
        for j in range(100):
            if i != j:
                W[i, j] += (vector_list[p][i] * vector_list[p][j])/100
            h_i += W[i,j] * network_state[j]
            s_i = sgn(h_i)
            network_state[i] = s_i
        # Check if the final network state is the same as the imprinted pattern
        if np.array_equal(network_state, vector_list[p]) and i <= p:
            stable_list[p] += 1

# calculate probability
for i in range(50):
    prob_list[i] = stable_list[i] / (i+1)
    unprob_list[i] = 1 - (stable_list[i] / (i+1))

print(stable_list)
print(prob_list)
print(unprob_list)
```
