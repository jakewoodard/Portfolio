```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_score
```

```python
# Read in CSV here
data = pd.read_csv("Life Expectancy Data.csv")

# Extract 2015
df = data.loc[data["Year"] == 2015]
print(df)
```

```python
# extract dataframes for life exp
developed_df = df.loc[df["Status"] == "Developed"]
developing_df = df.loc[df["Status"] == "Developing"]
```

```python
# Question 1.1
countries = np.unique(df["Country"])
Life_Ex = []
labels = ["Developed", "Developing"]
Life_E = np.array(developed_df["Life expectancy "])
Life_Ex.append(Life_E)
Life_E1 = np.array(developing_df["Life expectancy "])
Life_Ex.append(Life_E1)

# boxplot
plt.boxplot(Life_Ex)
plt.ylabel("Life Expectancy")
plt.xticks(range(1,len(labels)+1), labels)
plt.title("Life Expectancy for Developed vs Developing Countries")
```

```python
# Question 1.2
# the scatter plot for life expectancy vs. GDP and annotate
GDPs =  np.array(df["GDP"])
Life_Expectancy = np.array(df["Life expectancy "])
plt.scatter(Life_Expectancy, GDPs)
plt.title("Scatterplot: Life expectancy vs GDP")
plt.xlabel("Life expectancy")
plt.ylabel("GDP")
```

```python
# Question 1.3
# stacked histogram for BMI
BMI_1 = developed_df[" BMI "]
BMI_3 = developing_df[" BMI "]
plt.bar(Life_E, BMI_1, label="Developed")
plt.bar(Life_E1, BMI_3, label="Developing")
plt.legend()
plt.xlabel("Life expectancy")
plt.ylabel("Body Mass Index")
```

```python
# Question 1.4
# line plot that shows average life expectancy changing over 
# time across all nations
years = np.unique(data["Year"])
le_avg = []
le_max = []
le_min = []
le_std1 = []
le_std2 = []
le_std = []

for year in years:
    x = np.array(data[data["Year"] == year]["Life expectancy "])
    x = x[~np.isnan(x)]
    le_avg.append(np.average(x))
    le_max.append(np.max(x))
    le_min.append(np.min(x))
    le_std.append(np.std(x))

# find standard dev
le_std1 = np.subtract(le_avg, le_std)
le_std2 = np.add(le_avg, le_std)
    
plt.plot(years, le_avg)
plt.plot(years, le_max, linestyle = 'dotted')
plt.plot(years, le_min, linestyle = 'dotted')
plt.plot(years, le_std1, linestyle = 'dashed')
plt.plot(years, le_std2, linestyle = 'dashed')
plt.fill_between(years, le_avg, le_std1, alpha = 0.3)
plt.fill_between(years, le_avg, le_std2, alpha = 0.3)
plt.legend(["avg", "max", "min"])
plt.xlabel("Years")
plt.ylabel("Average Life Expectancy")
plt.title("Average Life Expectancy for All Nations over time")
```

```python
# Question 2.1


new_df = data[["Status", "Adult Mortality", "Hepatitis B", "Polio", "Measles ", " BMI ", "Diphtheria ", " HIV/AIDS"]]
new_df = new_df.dropna()
y = np.array(new_df["Status"])
X = np.array(new_df.drop(['Status'], axis=1))

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.33)

clf = tree.DecisionTreeClassifier(criterion="entropy")
clf.fit(X_train, y_train)
y_predict = clf.predict(X_train)
y_predict = clf.predict(X_test)
print(accuracy_score(y_test, y_predict))

# The accuracy score is 0.8807692307692307
# Body Mass Index is the first feature it used and
# it split it on 52.45
```

```python
# Question 2.2
# I will demonstrate this answer using
# max depth values of 2, 6, 10, 15, 20

y = np.array(new_df["Status"])
X = np.array(new_df.drop(['Status'], axis=1))

depth = [2, 6, 10, 15, 20]
train_acc = []
test_acc = []

# Max_depth = 2
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.33)

clf = tree.DecisionTreeClassifier(criterion="entropy", max_depth = 2)
clf.fit(X_train, y_train)
y_predict = clf.predict(X_train)
train_acc.append(accuracy_score(y_train, y_predict))

y_predict = clf.predict(X_test)
test_acc.append(accuracy_score(y_test, y_predict))


# Max_depth = 6
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.33)

clf = tree.DecisionTreeClassifier(criterion="entropy", max_depth = 6)
clf.fit(X_train, y_train)
y_predict = clf.predict(X_train)
train_acc.append(accuracy_score(y_train, y_predict))

y_predict = clf.predict(X_test)
test_acc.append(accuracy_score(y_test, y_predict))


# Max_depth = 10
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.33)

clf = tree.DecisionTreeClassifier(criterion="entropy", max_depth = 10)
clf.fit(X_train, y_train)
y_predict = clf.predict(X_train)
train_acc.append(accuracy_score(y_train, y_predict))

y_predict = clf.predict(X_test)
test_acc.append(accuracy_score(y_test, y_predict))


# Max_depth = 15
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.33)

clf = tree.DecisionTreeClassifier(criterion="entropy", max_depth = 15)
clf.fit(X_train, y_train)
y_predict = clf.predict(X_train)
train_acc.append(accuracy_score(y_train, y_predict))

y_predict = clf.predict(X_test)
test_acc.append(accuracy_score(y_test, y_predict))

# Max_depth = 20
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.33)

clf = tree.DecisionTreeClassifier(criterion="entropy", max_depth = 20)
clf.fit(X_train, y_train)
y_predict = clf.predict(X_train)
train_acc.append(accuracy_score(y_train, y_predict))

y_predict = clf.predict(X_test)
test_acc.append(accuracy_score(y_test, y_predict))

# create graph and table 
# print both
plt.bar(depth, train_acc)
plt.bar(depth, test_acc)
plt.xlabel("Max Depth")
plt.xlabel("Accuracy")
plt.xticks(depth)
plt.legend(['Train Accuracy', 'Test Accuracy'], loc = 'lower right')
plt.show()

print("TABLE")
print("Max Depth ---------- 2     6     10    15    20   ")
print("Training Accuracy -- ", end = "")
for num in train_acc:
    print(f'{num:.3f}', end = " ")

print()
print("Testing Accuracy --- ", end = "")
for num in test_acc:
    print(f'{num:.3f}', end = " ")

print()
```

```python
# Question 2.5

# first tree
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.33)

clf = tree.DecisionTreeClassifier(criterion="entropy", max_depth = 15, min_samples_leaf = 10, min_samples_split = 2)
clf.fit(X_train, y_train)
y_predict = clf.predict(X_train)
training1 = accuracy_score(y_train, y_predict)

y_predict = clf.predict(X_test)
testing1 = accuracy_score(y_test, y_predict)

# second tree
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.33)

clf = tree.DecisionTreeClassifier(criterion="entropy", max_depth = 15, min_samples_leaf = 10, min_samples_split = 30)
clf.fit(X_train, y_train)
y_predict = clf.predict(X_train)
training2 = accuracy_score(y_train, y_predict)

y_predict = clf.predict(X_test)
testing2 = accuracy_score(y_test, y_predict)

# print scores
print("Tree 1")
print("training accuracy = " + str(training1))
print("testing accuracy  = " + str(testing1))
print("Tree 2")
print("training accuracy = " + str(training2))
print("testing accuracy  = " + str(testing2))

# Tree number 2 performs better
```
