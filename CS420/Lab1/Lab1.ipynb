```python
# isolate class 4
df = data.loc[data["Class"] == 4]

print(df["Lambda"])

# averages first
avg_Lambda = np.mean(df['Lambda'])
avg_Lambda_t = np.mean(df['Lambda_t'])
avg_H = np.mean(df['H'])
avg_H_t = np.mean(df['H_t'])

# now standard deviations
std_Lambda = np.std(df['Lambda'])
std_Lambda_t = np.std(df['Lambda_t'])
std_H = np.std(df['H'])
std_H_t = np.std(df['H_t'])

print(avg_Lambda)
print(avg_Lambda_t)
print(avg_H)
print(avg_H_t)
print(std_Lambda)
print(std_Lambda_t)
print(std_H)
print(std_H_t)
```

```python
# adjust class values in set
# 3 -> 2
# 4 -> 1
# 1,2 -> 0
data = pd.read_csv("PureExperimentData.csv")
ph = 0

for i in range(len(data)):
    x = data.at[i,'Class']
    if (x == 1):
        ph = 0
    if (x == 2):
        ph = 0
    if (x == 3):
        ph = 2
    if (x == 4):
        ph = 1
    # update value 
    data.at[i,'Class'] = ph

#print(data)
```
